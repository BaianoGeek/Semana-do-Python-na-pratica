#### 2 - Carregar dados de um arquivo Excel

```python
import pandas as pd
import plotly.express as px
dados = pd.read_excel("vendas.xlsx")

#Cabecalho
dados.head()
#Rodape
#dados.tail()
```

#### 5 - Gráficos interativos

```python
grafico = px.histogram(dados, x="loja",
                              y= "preco",
                              color= "forma_pagamento",
                              text_auto = True,
                              title = "Faturamento por Loja")
grafico.show()
```
